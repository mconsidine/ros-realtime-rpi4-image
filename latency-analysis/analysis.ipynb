```python
import matplotlib
import matplotlib.pyplot as plt
import os.path
from cyclictest_latency_plotter import load_latency_hist, plot_latency_hist
```

```python
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.size'] = 12
matplotlib.rcParams['figure.dpi'] = 200
matplotlib.rcParams['savefig.dpi'] = 200
```

```python
data = load_latency_hist(
    "data/22.04_5.15.39-rt42-raspi.log",
    style="k-", # black (k) solid (-) line
    label="22.04 + 5.15.39-rt42-raspi",
)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plot_latency_hist([data], ax=ax, xlim=[0, 200])

# fig.savefig("data/{}.png".format(os.path.splitext(os.path.basename(data[1]["filename"]))[0]))
```
